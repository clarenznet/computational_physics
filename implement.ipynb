# Making a graph

```python
y=[]
x=[1,2,3,4,5]
for a in x:
    b=0
    b=a+3
    y.append(b)
# y=x[4]+3
print (y)

import matplotlib.pyplot as plt 
#plt(x,y)

# plot the data
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.plot(x, y, color='tab:orange')
#ax.plot(xdata2, ydata2, color='tab:orange')

# Add title and axis names
plt.title('graphing example')
plt.xlabel('original numbers')
plt.ylabel('add 3 for each x value given')
```

```python
import numpy as np
import math
time=np.linspace(0.0,10.0,100)
#print (time)
height2=time*3
#print (height2)
height=np.exp(-time/3.0)*np.sin(time*3)
print (height)

# plot the data
#fig = plt.figure()
#ax = fig.add_subplot(1, 1, 1)
#ax.plot( time,height, color='tab:red')
#ax.plot( time,0.3*np.sin(time*3), color='tab:blue')
plt.plot(time, height, "-r", label="damped")
plt.plot(time, 0.3*np.sin(time*3), "-b", label="constant amplitude")
plt.legend(loc="upper right")
plt.ylim(-1.5, 2.0)
plt.title('Damped and undamped signal')
plt.xlabel('Time (S)')
plt.ylabel('Amplitude')

plt.show()

#ax.plot(xdata2, ydata2, color='tab:orange')

# Add title and axis names

```

```python
x = np.linspace(0, 20, 1000)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, "-b", label="sine")
plt.plot(x, y2, "-r", label="cosine")
plt.legend(loc="upper left")
plt.ylim(-1.5, 2.0)
plt.show()
```
