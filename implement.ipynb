# Making a graph

```python
y=[]
x=[1,2,3,4,5]
for a in x:
    b=0
    b=a+3
    y.append(b)
# y=x[4]+3
print (y)

import matplotlib.pyplot as plt 
#plt(x,y)

# plot the data
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.plot(x, y, color='tab:blue')
#ax.plot(xdata2, ydata2, color='tab:orange')
```
